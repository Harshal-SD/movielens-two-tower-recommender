```python
import torch
import torch.nn as nn
```

```python
from collections import UserDict


class matrixFactorization(nn.Module):

    def __init__(self, num_users, num_movies, size_embedding):
        super().__init__()

        self.user_embeddings = nn.Embedding(
            num_users,
            embedding_dim = size_embedding
        )

        self.movie_embeddings = nn.Embedding(
             num_movies,
             embedding_dim = size_embedding
        )

    def forward(self, user_id, movie_id):

        user_vecs = self.user_embeddings(user_id)

        movie_vecs = self.movie_embeddings(movie_id)

        predictions = (user_vecs * movie_vecs).sum(dim = 1)

        return predictions
```

```python
rec_model = matrixFactorization(5000, 2500, 32)

lossfun = nn.MSELoss()

optimizer = torch.optim.Adam(rec_model.parameters(), lr = 0.001)




print(rec_model.user_embeddings.weight.shape, rec_model.movie_embeddings.weight.shape, end = '\n\n')

userId = torch.tensor( [15, 25, 80] )
movieId = torch.tensor( [100, 150, 200] )
ratings = torch.tensor( [5.0,3.0,4.0] )

print(rec_model.user_embeddings(userId).shape)
print(rec_model.movie_embeddings(movieId).shape)
print('\n\n\n\n')
predictions = rec_model(userId, movieId)

print(predictions)
# print()
```

```python
predictions = rec_model(
    userId,
    movieId
)

loss = lossfun(
    predictions,
    ratings
)

optimizer.zero_grad()

loss.backward()

optimizer.step()
```
